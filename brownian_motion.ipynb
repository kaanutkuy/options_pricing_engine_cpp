```python
import math
import itertools
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
M = 10
t = 10
```

```python
# create 10 random walk paths
random_walk = [-1,1]
steps = np.random.choice(random_walk, size=(M,t)).T
origin = np.zeros((1,M))
rw_paths = np.concatenate([origin, steps]).cumsum(axis=0)
```

```python
plt.plot(rw_paths)
plt.xlabel("years (t)")
plt.ylabel("move")
plt.show()
```

```python
# create quadratic variation and variance functions
q_variation = lambda x: round(np.square(x[:-1] - x[1:]).sum(), 3)
variance = lambda x: round(np.var(x, axis=0), 3)
```

```python
[q_variation(path) for path in rw_paths.T[0:4]] 
```

```python
[variance(path) for path in rw_paths[1:11]]
```

```python
M = 10
t = 10
n = 10
```

```python
# create a scaled random walk to approximate Brownian motion
# W^(n) (t) = (1/sqrt(n)) * M_nt
random_walk = [-1,1]
steps = (1/np.sqrt(n)) * np.random.choice(random_walk, size=(M,t*n)).T
origin = np.zeros((1,M))
srw_paths = np.concatenate([origin, steps]).cumsum(axis=0)
```

```python
time = np.linspace(0, t, t*n + 1)
tt = np.full(shape=(M, t*n+1), fill_value=time)
tt = tt.T

plt.plot(tt, srw_paths)
plt.xlabel("Years (t)")
plt.ylabel("Move")
plt.show()
```
